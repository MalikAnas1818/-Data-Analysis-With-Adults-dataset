# 17. Replace income values['<=50K' , '>50K'] With 0 and 1


```python
df['income'].unique()
```

```python
df['income'].value_counts()
```

```python
sns.countplot(x='income',data=df)
plt.show()
```

```python
def salary_data(sal):
    if sal=='<=50K':
        return 0
    else:
        return 1
```

```python
salary_data('<=50K')
```

```python
df['salary']=df['income'].apply(salary_data)
```

```python
df.head(1)
```

```python
df.shape[1]
```

```python
df1=df
```

```python
df1.replace(to_replace=['<=50K', '>50K'],value=[0,1],inplace=True)
```

```python
df1
```
